```python
# ===== Bài 3: Đổi tiền bằng quy hoạch động =====

def coin_change_min(amount: int, coins):
    """
    Trả về (số_xu_ít_nhất, danh_sách_xu). Không có nghiệm -> (inf, []).
    """
    coins = sorted(set(int(c) for c in coins if c > 0))
    if amount < 0: return float("inf"), []
    dp = [float("inf")] * (amount + 1)
    take = [-1] * (amount + 1)
    dp[0] = 0
    for a in range(1, amount + 1):
        best_c = -1
        best = float("inf")
        for c in coins:
            if c > a: break
            v = dp[a - c] + 1
            if v < best:
                best = v
                best_c = c
        dp[a] = best
        take[a] = best_c
    if dp[amount] == float("inf"):
        return float("inf"), []
    # truy vết
    res = []
    a = amount
    while a > 0:
        c = take[a]
        res.append(c)
        a -= c
    return dp[amount], res

def coin_change_count(amount: int, coins):
    """
    Đếm số cách (không phân biệt thứ tự) để tạo amount từ coins.
    """
    coins = sorted(set(int(c) for c in coins if c > 0))
    if amount < 0: return 0
    dp = [0] * (amount + 1)
    dp[0] = 1
    for c in coins:
        for a in range(c, amount + 1):
            dp[a] += dp[a - c]
    return dp[amount]


# ===== Bài 4: Dãy tăng dài nhất (O(n log n)) kèm truy vết =====

def lis_length_and_sequence(arr):
    """
    Trả về (độ_dài, một_dãy_LIS). Tăng nghiêm ngặt.
    """
    n = len(arr)
    if n == 0: return 0, []
    # lower_bound không dùng bisect
    def lower_bound(t, x):
        lo, hi = 0, len(t)
        while lo < hi:
            mid = (lo + hi) // 2
            if t[mid] < x:
                lo = mid + 1
            else:
                hi = mid
        return lo

    tails = []          # giá trị nhỏ nhất tại mỗi độ dài
    tails_idx = []      # chỉ số ứng với tails
    prev = [-1] * n     # truy vết

    for i, x in enumerate(arr):
        pos = lower_bound(tails, x)  # dùng < để đảm bảo "tăng" (không cho bằng)
        if pos == len(tails):
            tails.append(x)
            tails_idx.append(i)
        else:
            tails[pos] = x
            tails_idx[pos] = i
        if pos > 0:
            prev[i] = tails_idx[pos - 1]

    # khôi phục dãy
    k = tails_idx[-1]
    seq = []
    while k != -1:
        seq.append(arr[k])
        k = prev[k]
    seq.reverse()
    return len(seq), seq


# ===== Bài 5: Xâu con palindrome dài nhất (O(n^2), tâm mở rộng) =====

def longest_palindrome(s: str) -> str:
    """
    Trả về xâu con palindrome dài nhất của s.
    """
    n = len(s)
    if n <= 1: return s
    best_l = 0
    best_r = 0

    def expand(l, r):
        while l >= 0 and r < n and s[l] == s[r]:
            l -= 1; r += 1
        return l + 1, r - 1  # biên hợp lệ

    for i in range(n):
        l1, r1 = expand(i, i)       # độ dài lẻ
        if r1 - l1 > best_r - best_l:
            best_l, best_r = l1, r1
        l2, r2 = expand(i, i + 1)   # độ dài chẵn
        if r2 - l2 > best_r - best_l:
            best_l, best_r = l2, r2

    return s[best_l:best_r + 1]
```
